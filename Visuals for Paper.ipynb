```python
# number of qubits in purification
n = 3

qbits = QuantumRegister(2*n, 'qbits')
check1 = ClassicalRegister(2, 'match_1')
check2 = ClassicalRegister(2, 'match_2')
cbits = ClassicalRegister(2, 'result')

sys = QuantumCircuit(qbits, check1, check2, cbits)

# preparing bell-states
for i in range (0, 2*n, 2):
    sys.h(qbits[i])
    sys.cx(qbits[i], qbits[i+1])

sys.barrier()

# Deutsch's Correction
for i in range(0, n):
    sys.rx(pi/2, qbits[2*i])
    sys.rx(-pi/2, qbits[2*i+1])

sys.barrier()

# Bennett's Protocol 
for i in range(1, n):
    sys.cx(qbits[0], qbits[2*i])
sys.barrier()    

for i in range(1, n):
    sys.cx(qbits[1], qbits[2*i+1])
    
sys.barrier()

sys.measure([qbits[2], qbits[4]], [check1[0], check2[0]])
sys.measure([qbits[3], qbits[5]], [check1[1], check2[1]])

sys.barrier()

# bell-basis measurement to check fidelity
sys.cx(qbits[0], qbits[1])
sys.h(qbits[0])

sys.measure([qbits[0], qbits[1]], [cbits[0], cbits[1]])

style = {'cregbundle': True, 'dpi': 1600, 'margin':[0.67,0,0,0], 'fontsize':13, 'subfontsize':11}

sys.draw(output='mpl', style=style, plot_barriers=False, initial_state=True, with_layout=True, scale=1, filename=path+'circ_deutsch.png')
```
